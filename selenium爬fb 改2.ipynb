```python
## æœå°‹è²¼æ–‡é€£çµ

def FindLinks(url, n):
    Links = []
    driver.get(url)
    for i in range(n):
        time.sleep(5)
        #æ¯æ¬¡æ²å‹• 4000çš„å–®ä½
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')  # driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    # é€™è£¡æœƒè·³å‡ºè¦æˆ‘å€‘ç™»å…¥çš„å¤§ç•«é¢ï¼Œæ‰¾åˆ°ã€Œç¨å¾Œå†èªªã€çš„æŒ‰éˆ•ä¸¦é»æ“Š
    #driver.find_element_by_xpath('//a[@id="expanding_cta_close_button"]').click()
    soup = BeautifulSoup(driver.page_source)
    posts = soup.findAll('div', {'class':'clearfix y_c3pyo2ta3'})
    for i in posts:
        Links.append('https://www.facebook.com' + i.find('a',{'class':'_5pcq'}).attrs['href'].split('?',2)[0])
    return Links
```

```python
def CrawlComment(soup):
#    for i in Links:
    try:
        userContent = soup.find('div', {'class':'_5pbx userContent _3576'})
        post_data = [u.get_text().strip() for u in userContent if userContent is not None]
    except:
        userContent = ''
    try:
        feedback1 = soup.find('a', {'class':'_3hg- _42ft'})
        post_coment = [u.get_text().strip() for u in feedback1 if feedback1 is not None]
    except:
        feedback1 = '0'
    try:
        feedback2 = soup.find('span', {'class':'_355t _4vn2'})
        post_share = [u.get_text().strip() for u in feedback2 if feedback2 is not None]
    except:
        feedback2 = '0'
    return pd.DataFrame(
        data = {'CommentContent':post_data,
                 'post_coment':post_coment,
                 'post_share':post_share
                },
        columns = ['CommentContent','post_coment','post_share'])
```
